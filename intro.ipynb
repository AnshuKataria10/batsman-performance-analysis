```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('rohit sharma vs virat kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='Black')
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('rohit sharma vs virat kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],linestyle='dotted',marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],linestyle='dashed',marker='*')
plt.title('rohit sharma vs virat kohli')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.grid()
```
